```python
import pandas as pd
import numpy as np
import plotly.express as px

```

```python
vehicles_us = pd.read_csv('vehicles_us.csv')
```

```python
print(vehicles_us.info())
display(vehicles_us.head())
```

```python
vehicles_us['model_year'] = vehicles_us['model_year'].fillna(0).astype(int)
vehicles_us['model_year'] = np.where(vehicles_us['model_year'] == 0, 'Unknown', vehicles_us['model_year'])

print(vehicles_us.info())
display(vehicles_us.head())
```

```python
vehicles_us['model_year'] = vehicles_us['model_year'].fillna('Unknown')
display(vehicles_us.head())
```

```python
median_cylinders = vehicles_us['cylinders'].median()
vehicles_us['cylinders'] = vehicles_us['cylinders'].fillna(median_cylinders)
display(vehicles_us.head())
```

```python
median_odometer = vehicles_us['odometer'].mean()
vehicles_us['odometer'] = vehicles_us[ 'odometer'].fillna(median_odometer).round(0)
display(vehicles_us.head())
```

```python
vehicles_us['paint_color'] = vehicles_us['paint_color'].fillna('Unknown') 
display(vehicles_us.head())
```

```python
vehicles_us['is_4wd'] = vehicles_us['is_4wd'].fillna(0.0)
display(vehicles_us.head())
```

```python
fig = px.histogram(vehicles_us, x="odometer") # crear un histograma
fig.show() # crear gráfico de dispersión
```
